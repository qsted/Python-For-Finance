如果折现率为3.5％，那么净现值（NPV）是多少？ 净现值定义为所有收益的现值减去所有成本的现值。 如果现金流入具有正号而现金流出具有负号，则NPV可以方便地定义为所有现金流的现值之和。 一个未来值的现值可通过以下公式估算：

<Br>

假设我们今天投资100美元，明年投资30美元，那么从第二年年底开始的未来5年，每年年底的未来现金流入将分别为10美元，40美元，50美元，45美元和20美元：



```python
cash_flows=[-100,-30,10,40,50,45,20]
def npv_f (rate,cash_flows):
    total=0.0
    for i,s in enumerate(cash_flows):
        total+=s/(1+rate)**i
    return total

print(npv_f(0.05,cash_flows))

x=["a","b","z"]
for i, value in enumerate(x):
    print(i,value)
```

```python
def npv_excel (rate,cash_flows):
    total=0.0
    for i,s in enumerate(cash_flows):
        total+=s/(1+rate)**(i+1)
    return total
print(npv_excel(0.05,cash_flows))
```

### introduction to scipy
The following are a few examples based on the functions enclosed in the SciPy
module. The sp.npv() function estimates the present values for a given set of cash
flows with the first cash flow happening at time zero. The first input value is the
discount rate, and the second input is an array of all cash flows.

```python
import scipy as sp
cashflows=[-100,50,40,20,10,50]
x=sp.npv(0.1,cashflows)
print(x)
```

What is the monthly cash flow to pay off a mortgage of $250,000 over 30 years
with an annual percentage rate (APR) of 4.5 percent, compounded monthly? The
following code shows the answer:

```python
pay=sp.pmt(0.045/12,30*12,250000)
print(pay)
```

What is the present value (e.g., the initial investment) of an investment that needs to total $15692.93 after 10 years of saving $100 every month? Assume the interest rate is 5% (annually) compounded monthly.

```python
np.pv(0.05/12, 10*12, 100, 15692.93)
```

By convention, the negative sign represents cash flow out (i.e., money not available today). Thus, to end up with $15,692.93 in 10 years saving $100 a month at 5% annual interest, one’s initial deposit should also be $100.

```python
a = np.array((0.05, 0.04, 0.03))/12
np.pv(a, 10*12, -100, 15692.93)
```

In finance, we estimate
both arithmetic and geometric means, which are defined in the following formulas.





Assume that we have three numbers of a, b, and c. Then their arithmetic mean is (a+b+c)/3, while their geometric mean is (a*b*c)^(1/3). For three values of 2, 3, and 4, we have the following two means:

```python
print("arithmetic mean", (2+3+4)/3.)

geo_mean=(2*3*4)**(1./3)
round(geo_mean,4)
```

If n returns are given, the formula to estimate their arithmetic mean remains the
same. However, the geometric mean formula for returns is different, as shown here:


To estimate a geometric mean, the sp.prod() function would be applied. The
function gives us the products of all data items; see the following code:


```python
import scipy as sp
import numpy as np

ret=sp.array([0.1,0.05,0.02])
print("arithmetic mean",sp.mean(ret))
print("geo mean", pow(sp.prod(ret+1),1/len(ret))-1)
```

Actually, a simple Python function could be written with just two lines to calculate a
geometric mean for a set of given returns; see the following code:

```python
def getGeoMeanReturn(x):
    return pow(sp.prod(ret+1),1/len(x))-1
getGeoMeanReturn(ret)
```
